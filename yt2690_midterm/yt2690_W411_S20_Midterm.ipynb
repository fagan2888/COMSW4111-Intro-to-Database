````python
import logging


def run_q(sql, args=None, fetch=True, cur=None, conn=midterm_conn, commit=True):
    '''
    Helper function to run an SQL statement.

    This is a modification that better supports HW1. An RDBDataTable MUST have a connection specified by
    the connection information. This means that this implementation of run_q MUST NOT try to obtain
    a default connection.

    :param sql: SQL template with placeholders for parameters. Cannot be NULL.
    :param args: Values to pass with statement. May be null.
    :param fetch: Execute a fetch and return data if TRUE.
    :param conn: The database connection to use. This cannot be NULL, unless a cursor is passed.
        DO NOT PASS CURSORS for HW1.
    :param cur: The cursor to use. This is wizard stuff. Do not worry about it for now.
        DO NOT PASS CURSORS for HW1.
    :param commit: This is wizard stuff. Do not worry about it.

    :return: A pair of the form (execute response, fetched data). There will only be fetched data if
        the fetch parameter is True. 'execute response' is the return from the connection.execute, which
        is typically the number of rows effected.
    '''

    cursor_created = False
    connection_created = False

    try:

        if conn is None:
            raise ValueError("In this implementation, conn cannot be None.")

        if cur is None:
            cursor_created = True
            cur = conn.cursor()

        if args is not None:
            log_message = cur.mogrify(sql, args)
        else:
            log_message = sql

        logger.debug("Executing SQL = " + log_message)

        res = cur.execute(sql, args)

        if fetch:
            data = cur.fetchall()
        else:
            data = None

        # Do not ask.
        if commit == True:
            conn.commit()

    except Exception as e:
        if commit == True:
            conn.rollback()
        raise(e)

    return (res, data)
````

<u>Answer</u>

(1) online reference: [quora source](https://www.quora.com/What-is-the-difference-between-a-file-system-and-a-database)
1. A database management system coordinates both the physical and the logical access to the data, whereas a file-processing system coordinates only the physical access. For example, data integrity is automatically maintained by DBMS, but not in file system.
2. A database management system is designed to allow flexible access to data (i.e. queries), whereas a file-processing system is designed to allow pre-determined access to data (i.e. compiled programs). Hence, file system is not flexible as DBMS.
3. DBMS provide back up and recovery whereas data lost in file system can't be recovered. Or, users have to manually back up a file system.
4. A database management system is designed to coordinate multiple users accessing the same data at the same time. A file-processing system is usually designed to allow one or more programs to access different data files at the same time.

(2) online reference: [GTU MCA Course Site](https://sites.google.com/site/gtublog/sem2/2620003/responsibilitiesofdbms)
1. Interaction with the File Manager: if not discharged, nothing stored in the files can be retrieved.
2. Integrity Enforcement: if not discharged, data will be a mess.
3. Security Enforcement: if not discharged, unauthorized user may have the access the data they are not allowed. This may cause some privacy issue.
4. Backup and Recovery: if not discharged, data could be lost permanently.
5. Concurrency Control: if not discharged, for example, incorrect bank balances might be reflected due to simultaneous withdrawals and deposits, and so on.

    - Declarative programming language gives declaration about the system you are building. In a declarative language, you just set the command or order, and let it be on the system how to complete that order. You just need your result without digging into how it should be done. Some common declarative programming languages are SQL, regular expression, and etc.

3. online reference: [toptal](https://www.toptal.com/software/declarative-programming)
    - readability: a declarative programming language is more like a natural language. Hence, it is easier for non-programmer to use.

- The university is planning for next semester. The initial analysis is to look at all previous sections for all years and semesters and offer all of the sections in the coming semester in the same building, room and time slot. Restated, the university is planning on offering 15 sections in the coming semester.

- There is a conflict if two different sections have the same time_slot_id, build and room_number.

- Produce a table of the form: ```building, room_number, time_slot_id, no_of_assigned_sections, list_of_sections``` for the potential schedule if the university offers all 15 sections. The columns are:
    - ```building, room_number, time_slot_id``` have the obvious meaning.
    - ```no_of_assigned_sections``` is the number of sections assigned to the building, room_number and time_slot_id.
    - ```assigned_sections``` is a text string of the form "course_id_1 (sec_id_1), "course_id_2 (sec_id_2), ..." for sections assigned to the building, room, time_slot_id. An example might be something of the form: ```CS-319 (2),EE-181 (1)```
<a id = 'q10'></a>


<hr style="height:1px;">

<u>Answer</u>

````python
%sql mysql+pymysql://dbuser:dbuserdbuser@localhost/newbook
````

````python
%%sql 
# all overlap courses
SELECT 
    A.course_id, A.sec_id, A.building, A.room_number, A.time_slot_id,
    B.course_id AS b_course_id, B.sec_id AS b_sec_id, B.time_slot_id AS b_time_slot_id
FROM
    newbook.section AS A, newbook.section AS B
WHERE
    (A.time_slot_id = B.time_slot_id OR (A.time_slot_id = 'H' AND B.time_slot_id = 'C'))
    AND A.building = B.building
    AND A.room_number = B.room_number
    AND (A.course_id != B.course_id OR A.sec_id != B.sec_id)
````

````python
%sql DROP VIEW all_courses
````

````python
%%sql 
# all previous sections for all years and semesters 
# and offer all of the sections in the coming semester in the same building, room and time slot
CREATE VIEW `all_courses` AS 
SELECT DISTINCT
    A.course_id, A.sec_id, A.building, A.room_number, A.time_slot_id
FROM
    newbook.section AS A, newbook.section AS B
WHERE
    A.time_slot_id != B.time_slot_id
    AND A.building != B.building
    AND A.room_number != B.room_number
    AND (A.course_id != B.course_id OR A.sec_id != B.sec_id)
````

````python
%sql SELECT * FROM all_courses
````

````python
%sql DROP VIEW all_sections
````

````python
%%sql
CREATE VIEW `all_sections` AS 
SELECT building, room_number, T.time_slot_id, COUNT(*) as no_of_assigned_sections
FROM time_slot AS T
JOIN all_courses AS C
ON T.time_slot_id = C.time_slot_id 
GROUP BY building, room_number, T.time_slot_id
````

````python
%sql SELECT * FROM all_sections;
````

````python
%sql DROP TABLE IF EXISTS offer_courses;
````

````python
%%sql
CREATE TABLE `offer_courses` (
    `building` VARCHAR(15) NOT NULL,
    `room_number` VARCHAR(7) NOT NULL,
    `time_slot_id` VARCHAR(4) NOT NULL,
    `no_of_assigned_sections` int,
    `assigned_sections` VARCHAR(64),
    PRIMARY KEY (`building`, `room_number`, `time_slot_id`)
)
````

````python
%%sql 
INSERT INTO offer_courses(building, room_number, time_slot_id, no_of_assigned_sections, assigned_sections)
SELECT DISTINCT 
    S.building, S.room_number, S.time_slot_id, 
    S.no_of_assigned_sections, 
    CONCAT(C.course_id,'(', C.sec_id,')') AS assigned_sections
FROM all_sections AS S
JOIN (
    SELECT *
    FROM all_courses
    WHERE sec_id IN (
        SELECT MIN(sec_id) FROM all_courses
        GROUP BY building, room_number, time_slot_id
    )
) AS C
USING (building, room_number, time_slot_id);
````

````python
# function
'''
CREATE DEFINER=`dbuser`@`localhost` FUNCTION `create_customer_no`(c_name VARCHAR(128)) RETURNS varchar(12) CHARSET utf8mb4
    READS SQL DATA
BEGIN

    DECLARE result VARCHAR(12);
    DECLARE prefix VARCHAR(3);
    DECLARE count INT;
    
    SET prefix = LOWER(SUBSTR(c_name, 1, 3));
    SELECT COUNT(*) INTO count FROM customers WHERE customer_name LIKE(CONCAT(prefix, "%"));
    SET count = count + 1;
    
    SET result = CONCAT(prefix, '-', count);
    

RETURN result;
END
'''
````

````python
# INSERT trigger
'''
CREATE DEFINER=`dbuser`@`localhost` TRIGGER `customers_BEFORE_INSERT` 
BEFORE INSERT ON `customers` FOR EACH ROW BEGIN

    DECLARE customer_num VARCHAR(16);
    SET customer_num = create_customer_no(New.customer_name);
    SET New.customer_number = customer_num;
    
END
'''

# UPDATE trigger
'''
CREATE DEFINER=`dbuser`@`localhost` TRIGGER `customers_BEFORE_UPDATE` 
BEFORE UPDATE ON `customers` FOR EACH ROW BEGIN

    IF new.customer_number != old.customer_number THEN 
        SET new.customer_number = old.customer_number;
        
        SIGNAL SQLSTATE '03001'
            SET message_text = 'ignoring the attempt to change customer number',
                MYSQL_ERRNO = 1001;
    END IF;
    
END
'''
````
